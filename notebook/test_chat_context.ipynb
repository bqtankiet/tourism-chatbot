```python
queries = [
    "giá vé bà nà hills",
    "giờ mở cửa cầu rồng",
    "địa chỉ hồ hoàn kiếm",
    "có gì chơi ở temple of literature",
    "thời tiết đà nẵng hôm nay",
    "bạn là ai"
]

for q in queries:
    out = ctx.process_query(q)
    print('-' * 80)
    print('query      :', out['query'])
    print('topic      :', out['topic'])
    print('intents    :', out['intents'])
    print('top scores :', out['topic_scores'][:2])
```
